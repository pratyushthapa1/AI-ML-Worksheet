```python
# Task 1: Measurement
def convert_length(value, from_unit, to_unit):
    # Conversion factors
    METER_TO_FEET = 3.28084

    if from_unit == 'm' and to_unit == 'ft':
        return value * METER_TO_FEET
    elif from_unit == 'ft' and to_unit == 'm':
        return value / METER_TO_FEET
    else:
        raise ValueError("Unsupported length unit conversion. Please use 'm' or 'ft'.")

def convert_weight(value, from_unit, to_unit):
    # Conversion factors
    KG_TO_LBS = 2.20462

    if from_unit == 'kg' and to_unit == 'lbs':
        return value * KG_TO_LBS
    elif from_unit == 'lbs' and to_unit == 'kg':
        return value / KG_TO_LBS
    else:
        raise ValueError("Unsupported weight unit conversion. Please use 'kg' or 'lbs'.")

def convert_volume(value, from_unit, to_unit):
    # Conversion factors
    LITER_TO_GALLON = 0.264172

    if from_unit == 'L' and to_unit == 'gal':
        return value * LITER_TO_GALLON
    elif from_unit == 'gal' and to_unit == 'L':
        return value / LITER_TO_GALLON
    else:
        raise ValueError("Unsupported volume unit conversion. Please use 'L' or 'gal'.")

def run_converter():
    print("Welcome to the Unit Converter!")

    while True:
        print("\nChoose a conversion type:")
        print("1. Length (meters <-> feet)")
        print("2. Weight (kilograms <-> pounds)")
        print("3. Volume (liters <-> gallons)")
        print("4. Exit")

        choice = input("Enter your choice (1-4): ")

        if choice == '4':
            print("Exiting Unit Converter. Goodbye!")
            break

        conversion_function = None
        from_unit_options = []
        to_unit_options = []
        unit_prompt = ""

        if choice == '1':
            conversion_function = convert_length
            from_unit_options = ['m', 'ft']
            to_unit_options = ['m', 'ft']
            unit_prompt = "Enter the unit to convert FROM (m for meters, ft for feet): "
        elif choice == '2':
            conversion_function = convert_weight
            from_unit_options = ['kg', 'lbs']
            to_unit_options = ['kg', 'lbs']
            unit_prompt = "Enter the unit to convert FROM (kg for kilograms, lbs for pounds): "
        elif choice == '3':
            conversion_function = convert_volume
            from_unit_options = ['L', 'gal']
            to_unit_options = ['L', 'gal']
            unit_prompt = "Enter the unit to convert FROM (L for liters, gal for gallons): "
        else:
            print("Invalid choice. Please enter a number between 1 and 4.")
            continue

        try:
            value_str = input("Enter the value to convert: ")
            value = float(value_str)

            from_unit = input(unit_prompt).lower()
            if from_unit not in from_unit_options:
                print(f"Invalid 'from' unit. Please choose from {', '.join(from_unit_options)}.")
                continue

            to_unit = input(f"Enter the unit to convert TO ({' or '.join([u for u in to_unit_options if u != from_unit])}): ").lower()
            if to_unit not in to_unit_options or to_unit == from_unit:
                print(f"Invalid 'to' unit or same as 'from' unit. Please choose from {', '.join(to_unit_options)} and not '{from_unit}'.")
                continue

            result = conversion_function(value, from_unit, to_unit)
            print(f"{value} {from_unit} is equal to {result:.4f} {to_unit}.")

        except ValueError as e:
            print(f"Error: Invalid input. {e}")
        except Exception as e:
            print(f"An unexpected error occurred: {e}")

# Run the converter
run_converter()
```

```python
# Task 2: Mathematical Expression
def calculate_sum(numbers):
    if not numbers:
        raise ValueError("The list of numbers cannot be empty.")
    return sum(numbers)

def calculate_average(numbers):
    """
    Calculates the average of a list of numbers.

    Args:
        numbers (list): A list of numeric values.

    Returns:
        float: The average of the numbers.

    Raises:
        ValueError: If the list is empty.
    """
    if not numbers:
        raise ValueError("The list of numbers cannot be empty to calculate the average.")
    return sum(numbers) / len(numbers)

def find_maximum(numbers):
    """
    Finds the maximum value in a list of numbers.

    Args:
        numbers (list): A list of numeric values.

    Returns:
        float: The maximum value in the list.

    Raises:
        ValueError: If the list is empty.
    """
    if not numbers:
        raise ValueError("The list of numbers cannot be empty to find the maximum.")
    return max(numbers)

def find_minimum(numbers):
    """
    Finds the minimum value in a list of numbers.

    Args:
        numbers (list): A list of numeric values.

    Returns:
        float: The minimum value in the list.

    Raises:
        ValueError: If the list is empty.
    """
    if not numbers:
        raise ValueError("The list of numbers cannot be empty to find the minimum.")
    return min(numbers)

def run_math_operations():
    """
    Main function to run the mathematical operations program.
    Prompts the user for an operation and a list of numbers, then displays the result.
    Handles potential errors such as invalid input or empty lists.
    """
    print("Welcome to the Mathematical Operations Program!")

    while True:
        print("\nChoose an operation:")
        print("1. Sum")
        print("2. Average")
        print("3. Maximum")
        print("4. Minimum")
        print("5. Exit")

        choice = input("Enter your choice (1-5): ")

        if choice == '5':
            print("Exiting program. Goodbye!")
            break

        if choice not in ['1', '2', '3', '4']:
            print("Invalid choice. Please enter a number between 1 and 5.")
            continue

        try:
            numbers_str = input("Enter a list of numbers separated by spaces: ")
            if not numbers_str.strip():
                raise ValueError("No numbers entered. The list cannot be empty.")

            numbers = [float(x) for x in numbers_str.split()]

            result = None
            operation_name = ""

            if choice == '1':
                result = calculate_sum(numbers)
                operation_name = "Sum"
            elif choice == '2':
                result = calculate_average(numbers)
                operation_name = "Average"
            elif choice == '3':
                result = find_maximum(numbers)
                operation_name = "Maximum"
            elif choice == '4':
                result = find_minimum(numbers)
                operation_name = "Minimum"

            print(f"The {operation_name} of the numbers is: {result}")

        except ValueError as e:
            print(f"Error: Invalid input. {e}")
        except Exception as e:
            print(f"An unexpected error occurred: {e}")

# Run the program
run_math_operations()
```

```python
# 1. Flatten a Nested List:
def flatten(lst):
    flattened_list = []
    for item in lst:
        if isinstance(item, list):
            flattened_list.extend(flatten(item))
        else:
            flattened_list.append(item)
    return flattened_list

nested_list_1 = [[1, 2], [3, 4], [5]]
print(f"Original nested list: {nested_list_1}")
print(f"Flattened list: {flatten(nested_list_1)}")

nested_list_2 = [1, [2, [3, 4]], 5]
print(f"Original nested list: {nested_list_2}")
print(f"Flattened list: {flatten(nested_list_2)}")
```

```python
# 2. Accessing Nested List Elements:
def access_nested_element(lst, indices):
    current_element = lst
    for index in indices:
        if not isinstance(current_element, list) or not (0 <= index < len(current_element)):
            raise IndexError(f"Invalid index {index} for current element {current_element}")
        current_element = current_element[index]
    return current_element

nested_list = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
indices_1 = [1, 2]
element_1 = access_nested_element(nested_list, indices_1)
print(f"Original nested list: {nested_list}")
print(f"Element at indices {indices_1}: {element_1}")

indices_2 = [0, 1]
element_2 = access_nested_element(nested_list, indices_2)
print(f"Element at indices {indices_2}: {element_2}")

nested_list_deep = [1, [2, [3, 4]], 5]
indices_3 = [1, 1, 0]
element_3 = access_nested_element(nested_list_deep, indices_3)
print(f"Original nested list: {nested_list_deep}")
print(f"Element at indices {indices_3}: {element_3}")

try:
    access_nested_element(nested_list, [1, 3])
except IndexError as e:
    print(f"Error: {e}")

try:
    access_nested_element(nested_list, [3, 0])
except IndexError as e:
    print(f"Error: {e}")
```

```python
# 3. Sum of All Elements in a Nested List:
def sum_nested(lst):
    total_sum = 0
    for item in lst:
        if isinstance(item, list):
            total_sum += sum_nested(item)
        elif isinstance(item, (int, float)):
            total_sum += item
    return total_sum

nested_list_1 = [[1, 2], [3, [4, 5]], 6]
print(f"Original nested list: {nested_list_1}")
print(f"Sum of all elements: {sum_nested(nested_list_1)}")

nested_list_2 = [10, [20, 30], [40, [50]]]
print(f"Original nested list: {nested_list_2}")
print(f"Sum of all elements: {sum_nested(nested_list_2)}")

nested_list_3 = [1]
print(f"Original nested list: {nested_list_3}")
print(f"Sum of all elements: {sum_nested(nested_list_3)}")

nested_list_4 = []
print(f"Original nested list: {nested_list_4}")
print(f"Sum of all elements: {sum_nested(nested_list_4)}")
```

```python
# 4. Remove Specific Element from a Nested List:
def remove_element(lst, elem):
    new_list = []
    for item in lst:
        if isinstance(item, list):
            modified_sublist = remove_element(item, elem)
            if modified_sublist:
                new_list.append(modified_sublist)
        elif item != elem:
            new_list.append(item)
    return new_list

nested_list_1 = [[1, 2], [3, 2], [4, 5]]
elem_1 = 2
modified_list_1 = remove_element(nested_list_1, elem_1)
print(f"Original nested list: {nested_list_1}")
print(f"After removing {elem_1}: {modified_list_1}")

nested_list_2 = [1, [2, [3, 2, 4]], 2, 5]
elem_2 = 2
modified_list_2 = remove_element(nested_list_2, elem_2)
print(f"Original nested list: {nested_list_2}")
print(f"After removing {elem_2}: {modified_list_2}")

nested_list_3 = [[1, 1], [1, 1]]
elem_3 = 1
modified_list_3 = remove_element(nested_list_3, elem_3)
print(f"Original nested list: {nested_list_3}")
print(f"After removing {elem_3}: {modified_list_3}")

nested_list_4 = [[], [1, [2]]]
elem_4 = 1
modified_list_4 = remove_element(nested_list_4, elem_4)
print(f"Original nested list: {nested_list_4}")
print(f"After removing {elem_4}: {modified_list_4}")
```

```python
# 5. Find the Maximum Element in a Nested List:
def find_max(lst):
    max_val = float('-inf') # Initialize with negative infinity
    found_numeric = False

    for item in lst:
        if isinstance(item, list):
            try:
                nested_max = find_max(item)
                if nested_max is not None:
                    max_val = max(max_val, nested_max)
                    found_numeric = True
            except ValueError: #
                pass
        elif isinstance(item, (int, float)):
            max_val = max(max_val, item)
            found_numeric = True

    if not found_numeric:
        raise ValueError("The list is empty or contains no comparable numeric elements.")

    if max_val == float('-inf') and not found_numeric:
         raise ValueError("The list is empty or contains no comparable numeric elements.")

    return max_val

nested_list_1 = [[1, 2], [3, [4, 5]], 6]
print(f"Original nested list: {nested_list_1}")
print(f"Maximum element: {find_max(nested_list_1)}")

nested_list_2 = [10, [20, 300], [40, [50]]]
print(f"Original nested list: {nested_list_2}")
print(f"Maximum element: {find_max(nested_list_2)}")

nested_list_3 = [7]
print(f"Original nested list: {nested_list_3}")
print(f"Maximum element: {find_max(nested_list_3)}")

try:
    find_max([])
except ValueError as e:
    print(f"Error for empty list: {e}")

try:
    find_max([['a', 'b'], ['c']])
except ValueError as e:
    print(f"Error for non-numeric list: {e}")

nested_list_4 = [1, 'a', [2, 'b', [3]]]
print(f"Original nested list: {nested_list_4}")
print(f"Maximum numeric element: {find_max(nested_list_4)}")
```

```python
# 6. Count Occurrences of an Element in a Nested List:
def count_occurrences(lst, elem):
    """
    Counts how many times a specific element appears in a nested list.

    Args:
        lst (list): The nested list to search.
        elem (any): The element to count.

    Returns:
        int: The number of occurrences of the element.
    """
    count = 0
    for item in lst:
        if isinstance(item, list):
            count += count_occurrences(item, elem)
        elif item == elem:
            count += 1
    return count

nested_list_1 = [[1, 2], [2, 3], [2, 4]]
elem_1 = 2
print(f"Original nested list: {nested_list_1}")
print(f"Occurrences of {elem_1}: {count_occurrences(nested_list_1, elem_1)}")

nested_list_2 = [1, [2, [3, 2, 4]], 2, 5]
elem_2 = 2
print(f"Original nested list: {nested_list_2}")
print(f"Occurrences of {elem_2}: {count_occurrences(nested_list_2, elem_2)}")

nested_list_3 = ['a', ['b', ['a', 'c']], 'a']
elem_3 = 'a'
print(f"Original nested list: {nested_list_3}")
print(f"Occurrences of {elem_3}: {count_occurrences(nested_list_3, elem_3)}")

nested_list_4 = []
elem_4 = 1
print(f"Original nested list: {nested_list_4}")
print(f"Occurrences of {elem_4}: {count_occurrences(nested_list_4, elem_4)}")
```

```python
import numpy as np

A = np.array([[3, 4], [7, 8]])
B = np.array([[5, 3], [2, 1]])

print(f"Given matrices:\nA = \n{A}\nB = \n{B}")

# 1. Prove A.A⁻¹ = I.
inverse_A = np.linalg.inv(A)
product_A_invA = np.dot(A, inverse_A)
identity_matrix = np.eye(2)

print(f"\n1. Proving A.A⁻¹ = I:")
print(f"Inverse of A:\n{inverse_A}")
print(f"A . A⁻¹:\n{product_A_invA}")
print(f"Identity Matrix I:\n{identity_matrix}")
print(f"Is A . A⁻¹ approximately equal to I? {np.allclose(product_A_invA, identity_matrix)}")

# 2. Prove AB ≠ BA.
product_AB = np.dot(A, B)
product_BA = np.dot(B, A)

print(f"\n2. Proving AB ≠ BA:")
print(f"A . B:\n{product_AB}")
print(f"B . A:\n{product_BA}")
print(f"Is A . B equal to B . A? {np.array_equal(product_AB, product_BA)}")

# 3. Prove (AB)ᵀ = BᵀAᵀ
product_AB_T = np.transpose(product_AB) # or product_AB.T
transpose_B = np.transpose(B) # or B.T
transpose_A = np.transpose(A) # or A.T
product_BT_AT = np.dot(transpose_B, transpose_A)

print(f"\n3. Proving (AB)ᵀ = BᵀAᵀ:")
print(f"(A . B)ᵀ:\n{product_AB_T}")
print(f"Bᵀ . Aᵀ:\n{product_BT_AT}")
print(f"Is (A . B)ᵀ approximately equal to Bᵀ . Aᵀ? {np.allclose(product_AB_T, product_BT_AT)}")
```

### Solving System of Linear Equations:

```python
import numpy as np

# Given system of linear equations:
# 2x - 3y + z = -1
# x - y + 2z = -3
# 3x + y - z = 9

# Represent the equations in Matrix form AX = B
# A is the coefficient matrix
A_coeffs = np.array([
    [2, -3, 1],
    [1, -1, 2],
    [3, 1, -1]
])

# B is the constants vector
B_constants = np.array([-1, -3, 9])

print(f"Coefficient Matrix A:\n{A_coeffs}")
print(f"Constants Vector B: {B_constants}")

# Solve using Inverse Method (X = A⁻¹B)
# First, find the inverse of A
inverse_A_coeffs = np.linalg.inv(A_coeffs)
print(f"\nInverse of A:\n{inverse_A_coeffs}")

# Then, multiply A⁻¹ by B
X_inverse_method = np.dot(inverse_A_coeffs, B_constants)
print(f"\nSolution using Inverse Method (x, y, z): {X_inverse_method}")

# Verify the solution by plugging back into A_coeffs * X
print(f"Verification (A * X): {np.dot(A_coeffs, X_inverse_method)}")
print(f"Expected B: {B_constants}")

# Solve using np.linalg.solve function
X_solve_function = np.linalg.solve(A_coeffs, B_constants)
print(f"\nSolution using np.linalg.solve (x, y, z): {X_solve_function}")

# Verify if both methods yield the same result
print(f"Are solutions from both methods approximately equal? {np.allclose(X_inverse_method, X_solve_function)}")
```

```python
# 7. Flatten a List of Lists of Lists (Deep Flatten):
def deep_flatten(lst):

    flattened_list = []
    for item in lst:
        if isinstance(item, list):
            flattened_list.extend(deep_flatten(item))
        else:
            flattened_list.append(item)
    return flattened_list

nested_list_1 = [[[1, 2], [3, 4]], [[5, 6], [7, 8]]]
print(f"Original deeply nested list: {nested_list_1}")
print(f"Deeply flattened list: {deep_flatten(nested_list_1)}")

nested_list_2 = [1, [2, [3, [4, 5], 6]], 7, [8]]
print(f"Original deeply nested list: {nested_list_2}")
print(f"Deeply flattened list: {deep_flatten(nested_list_2)}")

nested_list_3 = []
print(f"Original deeply nested list: {nested_list_3}")
print(f"Deeply flattened list: {deep_flatten(nested_list_3)}")
```

```python
# 8. Nested List Average:
def average_nested(lst):
    total_sum = 0
    count = 0

    def _sum_and_count(inner_lst):
        nonlocal total_sum, count
        for item in inner_lst:
            if isinstance(item, list):
                _sum_and_count(item)
            elif isinstance(item, (int, float)):
                total_sum += item
                count += 1

    _sum_and_count(lst)

    if count == 0:
        raise ValueError("The nested list is empty or contains no numeric elements to average.")

    return total_sum / count

nested_list_1 = [[1, 2], [3, 4], [5, 6]]
print(f"Original nested list: {nested_list_1}")
print(f"Average of elements: {average_nested(nested_list_1)}")

nested_list_2 = [10, [20, 30], [40, [50]]]
print(f"Original nested list: {nested_list_2}")
print(f"Average of elements: {average_nested(nested_list_2)}")

nested_list_3 = [7]
print(f"Original nested list: {nested_list_3}")
print(f"Average of elements: {average_nested(nested_list_3)}")

nested_list_4 = [1, 'a', [2, 'b', [3, 'c']]]
print(f"Original nested list: {nested_list_4}")
print(f"Average of numeric elements: {average_nested(nested_list_4)}")

try:
    average_nested([])
except ValueError as e:
    print(f"Error for empty list: {e}")

try:
    average_nested([['a', 'b'], ['c']])
except ValueError as e:
    print(f"Error for non-numeric list: {e}")
```
